```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize']=10,6
```

```python
#Importing dataset
df= pd.read_csv(r'G:\Rhishi\singlerow.csv').fillna(0)
df.tail()
```

```python
df.Date=pd.to_datetime(df.Date)
df=df.set_index('Date')
df.tail()
```

```python
#plot graph
plt.xlabel("Date")
plt.ylabel("Views to webpage")
plt.plot(df)
```

```python
#determining rolling statistics
rolmean=df.rolling(window=365).mean()

rolstd=df.rolling(window=365).std()
print(rolmean,rolstd)
```

```python
#plot rolling statistics
orig=plt.plot(df,color='blue',label='original')
mean=plt.plot(rolmean,color='red',label='rolling mean')
std=plt.plot(rolstd,color='pink',label='rolling std')
plt.legend(loc='best')
plt.title("rolling mean and standard deviation")
plt.show()
```

```python
#perform Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller

print("Result of Dickey-Fuller Test:")
dftest=adfuller(df['Views'],autolag='AIC')

dfoutput=pd.Series(dftest[0:4],index=['test statistics','p-value','lags used','number of observation used'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s) %key']=value
    print(dfoutput)
```

```python
#Estimating trend
df_logScale=np.log(df)
plt.plot(df_logScale)
```

```python
movingAverage=df_logScale.rolling(window=365).mean()
movingstd=df_logScale.rolling(window=365).std()
orig=plt.plot(df,color='blue',label='original')
mean=plt.plot(movingAverage,color='red',label='rolling mean')
std=plt.plot(movingstd,color='green',label='rolling std')
plt.legend(loc='best')
plt.title("rolling mean and standard deviation")
plt.show()
```

```python
df_logScaleminusMA=df_logScale-movingAverage
df_logScaleminusMA.head(10)

#remove Nan value
df_logScaleminusMA.dropna(inplace=True)
df_logScaleminusMA.head(10)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    #determining rilling statistics
    movingaverage=timeseries.rolling(window=365).mean()
    movingstd=timeseries.rolling(window=365).std()
    
    #plot rolling statistics
    orig=plt.plot(timeseries,color='blue',label='original')
    mean=plt.plot(movingaverage,color='red',label='rolling mean')
    std =plt.plot( movingstd,color='pink',label='rolling std')
    plt.legend(loc='best')
    plt.title("rolling mean and standard deviation")
    plt.show()
    
    
    #adfuller test
    print("Result of Dickey-Fuller Test:")
    dftest=adfuller(df['Views'],autolag='AIC')

    dfoutput=pd.Series(dftest[0:4],index=['test statistics','p-value','lags used','number of observation used'])
    for key,value in dftest[4].items():
        dfoutput['critical value (%s) %key']=value
        print(dfoutput)
```

```python
test_stationarity(df_logScaleminusMA)
```

```python
exponentialdecayweightedaverage=df_logScale.ewm(halflife=365,min_periods=0,adjust=True).mean()
plt.plot(df_logScale)
plt.plot(exponentialdecayweightedaverage,color='red')
```

```python
df_logScaleMinusexponentialdecayweightedaverage=df_logScale-exponentialdecayweightedaverage
test_stationarity(df_logScaleMinusexponentialdecayweightedaverage)
```

```python
df_logdiffshifting=df_logScale-df_logScale.shift()
plt.plot(df_logdiffshifting)
```

```python
df_logdiffshifting.dropna(inplace=True)
test_stationarity(df_logdiffshifting)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition=seasonal_decompose(df_logScale,period=100)

trend=decomposition.trend
seasonal=decomposition.seasonal
residual=decomposition.resid

plt.subplot(411)
plt.plot(df_logScale,label='original')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(trend,label='Trend')
plt.legend(loc='best')
 
plt.subplot(413)
plt.plot(seasonal,label='seasonality')
plt.legend(loc='best')
           
plt.subplot(414)
plt.plot(residual,label='Residual')
plt.legend(loc='best')
            
           
```

```python
decomposedlogdata=residual
decomposedlogdata.dropna(inplace=True)
test_stationarity(decomposedlogdata)
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf,pacf

lag_acf=acf(df_logdiffshifting,nlags=10)
lag_pacf=pacf(df_logdiffshifting,nlags=10,method='ols')

#plot ACF
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_logdiffshifting)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_logdiffshifting)),linestyle='--',color='gray')
plt.title("Autocorrelation Function")
        
#plot PACF
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_logdiffshifting)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_logdiffshifting)),linestyle='--',color='gray')
plt.title("Partial Autocorrelation Function")            
plt.tight_layout()            
            
            
            
                           
```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR model
model=ARIMA(df_logScale,order=(0,1,2))
results_AR=model.fit(disp=1)
plt.plot(df_logdiffshifting)
plt.plot(results_AR.fittedvalues,color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-df_logdiffshifting['Views'])**2))
print("plotting AR model")
```

```python
#MA model
model=ARIMA(df_logScale,order=(2,1,0))
results_MA=model.fit(disp=1)
plt.plot(df_logdiffshifting)
plt.plot(results_MA.fittedvalues,color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-df_logdiffshifting['Views'])**2))
print("plotting MA model")
```

```python
#Arima model
model=ARIMA(df_logScale,order=(2,1,2))
results_ARIMA=model.fit(disp=1)
plt.plot(df_logdiffshifting)
plt.plot(results_ARIMA.fittedvalues,color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-df_logdiffshifting['Views'])**2))
print("plotting ARIMA model")
```

```python
predictions_ARIMA_diff=pd.Series(results_ARIMA.fittedvalues,copy=True)
print(predictions_ARIMA_diff.head())
```

```python
#convert to commulative sum
predictions_ARIMA_diff_cumsum=predictions_ARIMA_diff.cumsum()
print(predictions_ARIMA_diff_cumsum.head())
```

```python
predictions_ARIMA_log=pd.Series(df_logScale['Views'].ix[0],index=df_logScale.index)
predictions_ARIMA_log=predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA=np.exp(predictions_ARIMA_log)
plt.plot(df)
plt.plot(predictions_ARIMA)
```
